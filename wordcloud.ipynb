```python
import re

def merge_numbers_with_context(tokens):
    merged = []
    i = 0
    while i < len(tokens):
        token = tokens[i]
        next_token = tokens[i+1] if i+1 < len(tokens) else ""

        # กรณีปี เช่น "ปี 2550"
        if token == "ปี" and re.fullmatch(r"\d{4}", next_token):
            merged.append(token + " " + next_token)
            i += 2

        # กรณีเงิน เช่น "2550 บาท"
        elif re.fullmatch(r"\d+(\.\d+)?", token) and next_token in ["บาท", "ล้าน", "แสน", "หมื่น", "พัน", "ร้อย", "วินาที", "นาที", "ชั่วโมง", "วัน", "เดือน", "ปี"]:
            merged.append(token + " " + next_token)
            i += 2

        # กรณีเวลา เช่น "19.15 น."
        elif re.fullmatch(r"\d{1,2}\.\d{1,2}", token) and next_token in ["น.", "นาฬิกา"]:
            merged.append(token + " " + next_token)
            i += 2

        # ลบตัวเลขโดดๆที่ไม่เข้ากลุ่มข้างบน
        elif re.fullmatch(r"\d+", token):
            i += 1  # skip

        else:
            merged.append(token)
            i += 1

    return merged


tokens = ['ตั้ง', 'รางวัลนำจับ', '5', 'หมื่น', 'บาท', 'ไล่', 'ล่า', 'คนร้าย', 'สวม', 'ชุด', 'ไรเดอร์', 'ใช้', '"', 'ปืน', 'ปลอม', '"', 'บุก', 'เดี่ยว', 'ชิง', 'ทอง', 'รูปพรรณ', '"', 'ร้านทองเยาวราช', 'กรุงเทพฯ', '"', 'สาขา', 'อุดรธานี', 'ได้', 'ไป', 'กว่า', '1.1', 'ล้าน', 'หลัง', '18', 'วัน', 'ยัง', 'ไร้', 'วี่แวว', 'ผกก.', 'สภ.เมืองอุดรธานี', 'ระบุ', 'ถือ', 'ได้', 'ว่า', 'เป็น', 'โจร', 'มือ', 'อาชีพ', 'เมื่อ', 'เวลา', '10.30', 'น.', 'วัน', 'ที่', '14', 'พฤษภาคม', '2568', 'ที่', 'ห้อง', 'ประชุม', 'ชุด', 'สืบสวน', 'สภ.เมืองอุดรธานี', 'พ.ต.อ.พัฒนวงศ์ จันทร์พล', 'ผกก.', 'สภ.เมืองอุดรธานี', 'เรียก', 'ประชุม', 'ตำรวจ', 'สืบสวน', 'สภ.เมืองอุดรธานี', 'และ', 'ตำรวจ', 'กองกำกับ', 'การสืบสวน', 'ภ.จว.อุดรธานี', 'เพื่อ', 'ติดตาม', 'ความ', 'คืบหน้า', 'คดี', 'โจรไรเดอร์', 'ควง', 'ปืน', 'ปลอม', 'บุก', 'เดี่ยว', 'ชิง', 'ทอง', 'หนัก', '23', 'บาท', 'มูลค่า', 'กว่า', '1.1', 'ล้าน', 'บาท', 'ร้าน', 'ทองเยาวราชกรุงเทพ', 'ชั้น', '2', 'ห้างบิ๊กซี', 'สาขา', 'อุดรธานี', 'แยกสันตพล', 'บายพาสสกลนคร', 'ถนนนิตโย', 'เขตเทศบาลนครอุดรธานี', 'หนี', 'ลอยนวล', 'เหตุ', 'เกิด', 'เวลา', '19.15', 'น.', 'ของ', 'วัน', 'ที่', '26', 'เมษายน', '2568', 'ผ่าน', 'ไป', '18', 'วัน', 'ตำรวจ', 'ยัง', 'ตาม', 'จับ', 'คนร้าย', 'ไม่', 'ได้', 'หลัง', 'ประชุม', 'เสร็จ', 'พ.ต.อ.พัฒนวงศ์ จันทร์พล', 'ผกก.', 'สภ.เมืองอุดรธานี', 'ได้', 'เปิดเผย', 'ว่า', 'หลัง', 'จาก', 'เกิด', 'เหตุ', 'ตำรวจ', 'สืบสวน', 'สภ.เมืองอุดรธานี', 'กอง', 'กับ', 'การ', 'สืบสวน', 'ภ.จว.อุดรธานี', 'และ', 'ตำรวจ', 'สืบสวน', 'ภ.4', 'ร่วม', 'กัน', 'ออก', 'สืบสวน', 'หา', 'ตัว', 'คนร้าย', 'โดย', 'มี', 'การ', 'ประชุม', 'ความ', 'คืบหน้า', 'ร่วม', 'กัน', 'ทุก', 'วัน', 'ที่', 'ห้อง', 'ประชุม', 'ชุด', 'สืบสวน', 'สภ.เมืองอุดรธานี', 'ตั้ง', 'รางวัลนำจับ', '5', 'หมื่น', 'โจรชิง', 'ทอง', 'เมืองอุดรธานี', '18', 'วัน', 'ยัง', 'ไร้', 'วี่แวว', 'พบ', 'เป็น', 'โจร', 'มือ', 'อาชีพ', 'ส่วน', 'ความ', 'คืบหน้า', 'ของ', 'คดี', 'ทราบ', 'ว่า', 'ก่อน', 'ก่อ', 'เหตุ', 'คนร้าย', 'ได้', 'ขับ', 'รถ', 'วน', 'อยู่', 'ใน', 'ซอยเพชร', 'น้ำ', 'ใส', 'เส้นทาง', 'บ้าน', 'หนองใส', 'เกือบ', '1', 'ทุ่ม', 'ก็', 'ขับ', 'มา', 'เข้า', 'เส้นทาง', 'สี่', 'แยกหนองใส', 'เลี้ยว', 'ซ้าย', 'เลียบ', 'รอบ', 'เมือง', 'เข้า', 'สี่', 'แยก', 'บายพาสสกลนคร', 'หรือ', 'แยก', 'บิ๊กซี', 'เวลา', '19.03', 'น.', 'เลี้ยว', 'ขวา', 'เข้า', 'หน้า', 'บิ๊กซี', 'แล้ว', 'จอด', 'จยย.', 'ฮอนด้า', 'คลิก', 'สี', 'เหลือง', 'ไว้', 'ที่', 'ถนน', 'หน้า', 'ห้าง', 'แล้ว', 'เดิน', 'เข้า', 'ประตู', 'ห้าง', 'ฝั่ง', 'ขวา', 'มือ', 'ขึ้น', 'บันได', 'เลื่อน', 'ไป', 'ก่อ', 'เหตุ', 'โดย', 'ใช้', 'เวลา', 'ก่อ', 'เหตุ', 'ประมาณ', '1', 'นาที', 'ค่อนข้าง', 'รวดเร็ว', 'ชิงเอา', 'ทอง', 'ไป', '23', 'บาท', 'ถือ', 'ได้', 'ว่า', 'เป็น', 'โจร', 'มือ', 'อาชีพ', 'และ', 'คล่อง', 'ตัว', 'พ.ต.อ.พัฒนวงศ์', 'เปิดเผย', 'ต่อ', 'ว่า', 'คนร้าย', 'ได้', 'ใช้', 'เส้นทาง', 'หลบหนี', 'โดย', 'ขี่', 'รถ', 'จักรยาน', 'ยนต์', 'ฮอนด้า', 'คลิก', 'สี', 'เหลือง', 'ไม่', 'ติด', 'แผ่น', 'ป้าย', 'ทะเบียน', 'ซึ่ง', 'เป็น', 'รถ', 'รุ่น', 'ปี', '2020', 'เลี้ยว', 'ซ้าย', 'หลบหนี', 'เข้า', 'ไป', 'ทาง', 'ตัวเมือง', 'อุดรธานี', 'แต่', 'ไป', 'จอด', 'ติด', 'ขบวน', 'รถไฟ', 'ที่', 'ทาง', 'พาด', 'รถไฟ', 'ตลาดหนองบัวแบบ', 'ใจเย็น', 'จาก', 'นั้น', 'ได้', 'เลี้ยว', 'ขวา', 'สาม', 'แยก', 'เอกอุดร', 'ไป', 'ถนน', 'ทองใหญ่', 'ผ่าน', 'ศูนย์', 'การ', 'ค้ายูดี', 'ทาวน์', 'โรงเรียนเทศบาล', '7', 'ตรง', 'ไป', 'ที่', 'ชุมชน', 'รถไฟ', 'แล้ว', 'ข้าม', 'ทาง', 'รถไฟ', 'ด้วย', 'ทาง', 'จักรยาน', 'ยนต์', 'ไป', 'ชุมชน', 'หนองเหล็ก', 'ผ่าน', 'วัดป่าหนองแก้ว', 'เลี้ยว', 'ซ้าย', 'ออก', 'ถนน', 'รอบ', 'เมือง', 'เลี้ยว', 'ซ้าย', 'ผ่าน', 'โฮมโปร', 'คนร้าย', 'จอด', 'รถ', 'แล้ว', 'เปลี่ยน', 'เสื้อ', 'เป็น', 'เสื้อ', 'แจ็กเก็ต', 'สี', 'ดำ', 'ตั้ง', 'รางวัลนำจับ', '5', 'หมื่น', 'โจรชิง', 'ทอง', 'เมืองอุดรธานี', '18', 'วัน', 'ยัง', 'ไร้', 'วี่แวว', 'พบ', 'เป็น', 'โจร', 'มือ', 'อาชีพ', '“', 'เมื่อ', 'คนร้าย', 'เปลี่ยน', 'เสื้อ', 'เสร็จ', 'ก็', 'ขี่', 'รถ', 'มุ่ง', 'ตรง', 'ไป', 'ถึง', 'ถนนสี่', 'แยกอุโมงค์', 'และ', 'ถนน', 'ต่าง', 'ระดับ', 'เลี้ยว', 'ขวา', 'ไป', 'ถนนมิตรภาพ', 'มุ่งหน้า', 'ไป', 'จ.หนองคาย', 'แล้ว', 'ไป', 'เลี้ยว', 'ซ้าย', 'ถนน', 'มัจฉา', 'ผ่าน', 'บ้านนา', 'กว้าง', 'ทะลุ', 'สี่', 'แยกเชียงพัง', 'เลี้ยว', 'ซ้าย', 'ไป', 'อ.บ้านผือ', 'ถึง', 'บ้านดู่', 'ต.บ้านขาว', 'อ.เมืองอุดรธานี', 'ซึ่ง', 'เป็น', 'หมู่', 'บ้าน', 'สุดท้าย', 'ของ', 'เขต', 'อำเภอเมือง', 'ได้', 'เลี้ยว', 'ขวา', 'ลง', 'ไป', 'ถนน', 'ใน', 'หมู่', 'บ้าน', 'ซึ่ง', 'สามารถ', 'ทะลุ', 'ไป', 'อ.สระใคร', 'จ.หนองคาย', 'ได้', 'จาก', 'นั้น', 'ก็', 'ได้', 'หาย', 'ไป', 'ไม่', 'สามารถ', 'ติดตาม', 'ได้', 'จึง', 'ประสาน', 'ตำรวจสืบสวน', 'ภ.จว.อุดรธานี', 'สืบสวน', 'ภาค', '4', 'ได้', 'ให้', 'ตำรวจ', 'ซึ่ง', 'มี', 'เขต', 'ติดต่อ', 'กัน', 'พร้อม', 'กับ', 'ประสาน', 'กำนัน', 'ผู้', 'ใหญ่', 'บ้าน', 'ปู', 'พรม', 'ค้นหา', 'คนร้าย', 'แต่', 'ก็', 'ยัง', 'ไม่', 'พบ', 'ตัว', 'พ.ต.อ.พัฒนวงศ์', 'เปิดเผย', 'อีก', 'ว่า', 'สำหรับ', 'ไรเดอร์', 'ตำรวจ', 'ได้', 'ทำ', 'หนังสือ', 'สอบถาม', 'ว่า', 'ใน', 'พื้นที่', 'อุดรธานี', 'ที่', 'มี', 'ผู้', 'เข้า', 'สมัคร', 'จด', 'ทะเบียน', 'โดย', 'ใช้', 'รถ', 'จักรยาน', 'ยนต์', 'ยี่ห้อ', 'ฮอนด้า', 'คลิก', 'สี', 'เหลือง', 'ก็', 'ให้', 'ส่ง', 'ข้อมูล', 'ให้', 'ทาง', 'สภ.เมืองอุดรธานี', 'ซึ่ง', 'ได้', 'ข้อมูล', 'มา', 'บาง', 'ส่วน', 'เมื่อ', 'เข้า', 'ตรวจสอบ', 'แล้ว', 'ไม่', 'พบ', 'พิรุธ', 'ส่วน', 'กลุ่ม', 'เป้าหมาย', 'อื่น', 'ก็', 'จะ', 'เป็น', 'บุคคล', 'พ้น', 'โทษ', 'ก่อ', 'คดี', 'ใน', 'พื้นที่', 'จ.อุดรธานี', 'มี', 'ทั้งหมด', '8,000', 'คน', 'ซึ่ง', 'ได้', 'เข้า', 'พิสูจน์', 'ทราบ', 'จยย.', 'ฮอนด้า', 'คลิก', 'สี', 'เหลือง', 'รุ่น', '2020', 'รวม', 'ทั้ง', 'ญาติพี่น้อง', 'ที่', 'ใช้', 'รถ', 'รุ่น', 'นี้', 'ทยอย', 'เข้า', 'ตรวจสอบ', 'และ', 'ประสาน', 'พื้นที่', 'ใกล้เคียง', 'อุดรธานี', 'หนองคาย', 'หนองบัวลำภู', 'เสื้อ', 'แกร็บ', 'และ', 'หมวก', 'กัน', 'น็อก', 'จะ', 'เป็น', 'ยี่ห้อ', 'ของ', 'เวียดนาม', 'ซื้อขาย', 'ทาง', 'ออนไลน์', 'ส่วน', 'ผู้', 'ต้อง', 'สงสัย', '4', 'ราย', 'ได้', 'นำ', 'มา', 'ตรวจ', 'ดีเอ็นเอ', 'เทียบ', 'กับ', 'คนร้าย', 'ปรากฏ', 'ว่า', 'ไม่', 'ตรง', 'กับ', 'คนร้าย', 'ตั้ง', 'รางวัลนำจับ', '5', 'หมื่น', 'โจรชิง', 'ทอง', 'เมืองอุดรธานี', '18', 'วัน', 'ยัง', 'ไร้', 'วี่แวว', 'พบ', 'เป็น', 'โจร', 'มือ', 'อาชีพ', '“', 'อยาก', 'ให้', 'ผู้', 'ที่', 'ขับ', 'รถ', 'ยนต์', 'และ', 'มี', 'กล้อง', 'หน้า', 'รถ', 'ขณะ', 'จอด', 'ติด', 'ขบวน', 'รถไฟ', 'ที่', 'ทาง', 'พาด', 'รถไฟ', 'ตลาดหนองบัว', 'เปิด', 'ดู', 'หาก', 'พบ', 'ภาพ', 'คนร้าย', 'ขอ', 'ให้', 'ส่ง', 'มา', 'ให้', 'ตำรวจ', 'สภ.เมืองอุดรธานี', 'และ', 'หาก', 'ประชาชน', 'ท่าน', 'ใด', 'มี', 'ความ', 'สงสัย', 'ว่า', 'บุคคล', 'ใน', 'ภาพ', 'ใน', 'คลิป', 'ที่', 'ก่อ', 'เหตุ', 'คล้าย', 'คน', 'ที่', 'รู้จัก', 'มี', 'เครื่อง', 'แต่ง', 'กาย', 'มี', 'รถ', 'มี', 'หมวก', 'กัน', 'น็อก', 'ที่', 'สงสัย', 'สามารถ', 'แจ้ง', 'ให้', 'ข้อมูล', 'มา', 'ยัง', 'พ.ต.ท.บรรจง พาโคตร', 'สว.', 'สส.', 'สภ.เมืองอุดรธานี', 'หรือ', 'พ.ต.อ.พัฒนวงศ์ จันทร์พล', 'ผกก.', 'สภ.เมืองอุดรธานี', 'หรือ', 'ชุด', 'สืบสวน', 'ที่', 'รู้จัก', 'เพื่อ', 'เป็น', 'ข้อมูล', 'นำ', 'ไป', 'สู่', 'การ', 'จับกุม', 'คนร้าย', 'ทาง', 'ร้านทอง', 'มี', 'รางวัลนำจับ', 'ให้', '50000', 'บาท', '”']
print(merge_numbers_with_context(tokens))
```
